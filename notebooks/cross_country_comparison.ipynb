```python
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "src")))

from eda_utils import SolarDataAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
benin = SolarDataAnalyzer('data/benin_clean.csv')
sierra = SolarDataAnalyzer('data/sierra_leone_clean.csv')
togo = SolarDataAnalyzer('data/togo_clean.csv')
```

```python
summary_df = pd.DataFrame({
    "Benin": benin.df.describe().T["mean"],
    "Sierra Leone": sierra.df.describe().T["mean"],
    "Togo": togo.df.describe().T["mean"]
})
summary_df
```

```python
ghi_df = pd.DataFrame({
    "GHI": pd.concat([
        benin.df["GHI"],
        sierra.df["GHI"],
        togo.df["GHI"]
    ]),
    "Country": (
        ["Benin"] * len(benin.df) +
        ["Sierra Leone"] * len(sierra.df) +
        ["Togo"] * len(togo.df)
    )
})

plt.figure(figsize=(10, 5))
sns.boxplot(data=ghi_df, x="Country", y="GHI")
plt.title("Comparison of GHI Across Countries")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(benin.df["GHI"], label="Benin", shade=True)
sns.kdeplot(sierra.df["GHI"], label="Sierra Leone", shade=True)
sns.kdeplot(togo.df["GHI"], label="Togo", shade=True)
plt.title("GHI Distribution Comparison")
plt.legend()
plt.show()
```

###  Observations
- Benin has a relatively **higher average GHI** with broader variability.
- Sierra Leones GHI is tightly clustered around the median.
- Togo shows intermediate GHI levels with more consistent spread.
- These trends may reflect cloud cover patterns or topographical influence.
