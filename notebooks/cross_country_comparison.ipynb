```python
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "src")))

from eda_utils import SolarDataAnalyzer
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
benin = SolarDataAnalyzer('data/benin_clean.csv')
sierra = SolarDataAnalyzer('data/sierra_leone_clean.csv')
togo = SolarDataAnalyzer('data/togo_clean.csv')
```

```python
import pandas as pd

# Create summary stats for each country
benin_stats = benin.summary_stats().loc[['mean', 'std']]
sierra_stats = sierra.summary_stats().loc[['mean', 'std']]
togo_stats = togo.summary_stats().loc[['mean', 'std']]

# Combine them for comparison
combined_stats = pd.concat(
    [benin_stats, sierra_stats, togo_stats],
    keys=['Benin', 'Sierra Leone', 'Togo']
)

combined_stats
```
